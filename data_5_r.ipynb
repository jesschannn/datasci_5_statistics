# T-Test

```r
df2 = read.csv('breast_cancer.csv')
df2
```

```r
malignant_data <- subset(df2, diagnosis == 'M')$radius_mean
malignant_data
```

```r
benign_data <- subset(df2, diagnosis == 'B')$radius_mean
benign_data
```

```r
t_test_result <- t.test(malignant_data, benign_data, var.equal = FALSE)
t_test_result
```
